```python
customer = pd.read_table('customers.csv', sep='\t', index_col=0)
customer.head()
```

```python
customer.isnull().any()
```

```python
customer.fillna(customer['Income'].mean(), inplace=True)
```

```python
customer.isnull().any()
```

```python
customer['total spending'] = customer['MntWines'] + customer['MntFruits'] + customer['MntMeatProducts'] + customer['MntFishProducts'] +customer['MntSweetProducts'] + customer['MntGoldProds'] 
customer.head()
```

```python
customer.groupby('Education')['total spending'].mean()
```

```python
customer.groupby(['Education','Marital_Status'])['Income'].mean()
```

```python
mask = (customer['Education']=='Master') | (customer['Education'] == 'PhD')
customer[mask].sort_values(by='total spending', ascending=False).head(10)
```

```python
customer['Dt_Customer'] = pd.to_datetime(customer['Dt_Customer'], dayfirst=True)
customer['Dt_Customer'].dtype
```

```python
customer.groupby(customer['Dt_Customer'].dt.year).size()
```

```python
customer.pivot_table('NumWebPurchases', index='Education',columns='Marital_Status').fillna('NotAvailable')
```

```python
pd.crosstab(customer['Education'],customer['Complain'])
```

```python
crime = pd.read_csv('HoustonCrime.csv', index_col='OBJECTID')
crime.head()
```

```python
crime.isnull().any()
```

```python
crime.dropna(subset=['Time_Begun','Zip_Code'], inplace = True)
```

```python
crime.isnull().any()
```

```python
crime['Zip_code']=crime['Zip_Code'].astype(int).astype('str')
```

```python
crime['Zip_code'].dtype
```

```python
crime['Time_Begun']=pd.to_datetime(crime['Time_Begun'])
crime['Time_Begun'] =crime['Time_Begun'].dt.date
crime.head()
```

```python
crime.groupby(['Time_Begun','Offense']).size()
```

```python
auto = (crime[crime['Offense']=='Auto Theft'])
auto.groupby('Time_Begun').size().sort_values(ascending=False).head(1)
```

```python
safe = crime.groupby('Zip_Code').size().sort_values().head()
safe
```

```python
mask= (crime['Offense']=='Murder') | (crime['Offense']=='Rape')
murder_rape = crime[mask]
murder_rape.groupby(['Offense','Time_Begun','Zip_Code']).size()
```

```python
crime['Time_Begun']= pd.to_datetime(crime['Time_Begun'])
pd.crosstab(crime['Offense'],crime['Time_Begun'].dt.month)
```

```python
crime.pivot_table('Offense',index='Time_Begun', columns='Zip_Code',aggfunc='count')
```

```python
nba = pd.read_csv('nba.csv')
```

```python
nba.isnull().any()
```

```python
nba['Date'].dtype
```

```python
nba['Date']=pd.to_datetime(nba['Date'])
nba['Date']=nba['Date'].dt.year
nba.head()
```

```python
 d = input('Please enter a team:')
```

```python
mask = nba['Team']==d
nba[mask].groupby(['Date','Home'])['TeamPoints'].mean()
```

```python
_2017 = nba[nba['Date']==2017]
pd.crosstab(_2017['Team'],_2017['WINorLOSS'])
```

```python
nba.pivot_table('TeamPoints', index='Team', columns='Date')
```

```python
_2017['Diff']= _2017['TeamPoints']-_2017['OpponentPoints']
```

```python
_2017.groupby('Team')['Diff'].mean().sort_values(ascending=False)
```

```python
nba['Diff'] = nba['TeamPoints']-nba['OpponentPoints']
nba.pivot_table('Diff', index='Team', columns='Date',aggfunc=['min','max'])
```
