```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
superstore = pd.read_excel('superstoreDataset.xlsx', index_col = 0)
superstore
```

```python
superstore.isnull().any()
```

```python
superstore[superstore['Product Base Margin'].isnull()]
```

```python
superstore.groupby(['State','Product Category','Product Sub-Category'])['Sales'].sum()
```

```python
state = superstore[superstore['State'] == 'Texas']
data = state.groupby('Product Sub-Category')['Profit'].sum()
data
```

```python
ax = data.plot.bar()
ax.set_title('Texas sales Profit analysis', fontsize = 12)
ax.set_xlabel('Product Sub-Category', fontsize = 12)
ax.set_ylabel('Profit')
```

```python
superstore['Order Date'].dtype
```

```python
dt = superstore.pivot_table('Sales', index = superstore['Order Date'].dt.year, columns = 'State', aggfunc='sum')
dt
```

```python
ax = dt.plot(marker = 's',figsize = (10,6))
ax.set_title('Total sales of each state from 2009 to 2012', fontsize = 15)
ax.set_xlabel('Year', fontsize = 15)
ax.set_ylabel('Total Sales', fontsize = 15)
ax.set_xticks([2009,2010,2011,2012])
```

```python
dt1 = superstore[superstore['Product Sub-Category']=='Appliances']['Unit Price']
ax = dt1.plot.hist(bins = 50)
ax.set_title('Appliances', fontsize=15)
ax.set_xlabel('Unit Price', fontsize = 15)
```

```python
ax = dt1.plot.density()
ax.set_title('Appliances', fontsize=15)
ax.set_xlabel('Unit Price', fontsize = 15)
```

```python
import seaborn as sns
ax = sns.distplot(dt1, bins=50)
ax.set_title('Appliances per hist and distribution density')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nba = pd.read_csv('nba.csv', index_col = 0)
nba
```

```python
nba.isnull().any()
```

```python
nba.columns
```

```python
nba['Date'].dtype
```

```python
nba['Date'] = pd.to_datetime(nba['Date'])
nba['Date'].dtype
```

```python
nba.groupby(['Team',nba['Date'].dt.year])['TeamPoints','OpponentPoints'].mean()
```

```python
nba_2017 = nba[nba['Date'].dt.year == 2017]
nba_2017
```

```python
data = nba_2017.groupby('Team')['TeamPoints'].mean()
ax = data.plot.bar()
ax.set_title('Average TeamPoints for all Teams', fontsize = 12)
ax.set_xlabel('Teams', fontsize = 12)
ax.set_ylabel('Average TeamPoints',fontsize = 12)
```

```python
nba_2017.groupby('Team')['OpponentPoints'].mean().sort_values().head()
```

```python
dt = pd.crosstab(nba_2017['Team'],nba_2017['WINorLOSS'])
ax = dt.plot.barh(stacked = True, figsize = (10,6))
ax.set_xlabel('WinorLoss')
ax.set_ylabel('Team')
```

```python
nba_2017['Diff'] = nba_2017['TeamPoints'] - nba_2017['OpponentPoints']
```

```python
nba_2017.groupby('Team')['Diff'].mean().sort_values(ascending = False)
```

```python
hou = nba_2017[nba_2017['Team'] == 'HOU']
ax = plt.scatter(hou['TeamPoints'], hou['OpponentPoints'])
plt.title('Scatter Plot')
```
